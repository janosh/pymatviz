```python
from pymatviz.ptable.plotly import ptable_bars_plotly
from smact.utils.oxidation import ICSD24OxStatesFilter
```

```python
ox_filter = ICSD24OxStatesFilter()
ox_states_df = ox_filter.ox_states_df
```

```python
df_nested_dict = (
    ox_states_df.groupby("element")
    .apply(
        lambda g: dict(
            zip(g["oxidation_state"], g["results_count"] / g["results_count"].sum())
        )
    )
    .to_dict()
)
```

```python
fig_intro = ptable_bars_plotly(
    df_nested_dict,
    colorscale="turbo",
    color_elem_strategy="both",  # Color both symbols and backgrounds
    elem_type_colors=None,  # Uses default element type colors
    scale=1,
    colorbar=False,
)
fig_intro
```
